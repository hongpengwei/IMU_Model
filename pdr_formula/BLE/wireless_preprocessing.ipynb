Input: camera2video BLE.csv,
Output: wireless.csv

```python
import pandas as pd
import numpy as np
```

```python
folder = '/content/drive/MyDrive/paper/0427/0427-6'
filepath = folder + '/BLE.csv'
df = pd.read_csv(filepath)
```

```python
col = ['label', 'Beacon_1', 'Beacon_2', 'Beacon_3', 'Beacon_4', 'Beacon_5', 'Beacon_6', 'Beacon_7', 'Time']
time = []
for i in range(df.shape[0]):
  t = df.iloc[i][0]
  time.append(t)
time = list(dict.fromkeys(time))

rows = []
for t in time:
  rssi = np.zeros(7)
  rssi[0:7] = np.nan
  mask = df['Time'] == t
  temp = df[mask] # 時間為 t 時的所有 beacon rssi
  for x in range(len(temp)):
    s = temp.iloc[x][1]
    if s[0:30] == 'e20a39f4-73f5-4bc4-a12f-17d1ad' :
      uuid_num = int(s[-1]) # last char of UUID
      rssi[uuid_num - 1] = temp.iloc[x][2]
  # print(rssi)
  rows.append([0, rssi[0], rssi[1], rssi[2], rssi[3], rssi[4], rssi[5], rssi[6], t])
# print(rows)
```

```python
rows = pd.DataFrame(rows, columns = col)
#rows[1:7] = rows[1:7].astype(float) # object to float for interpolate function
rows = rows.interpolate(limit_direction='both')
rows.set_index('label').to_csv(folder + '/wireless.csv')
```
