```python
import pandas as pd 
import numpy as np
import csv
from datetime import datetime
import pickle
from random import seed
from random import random
import math
import os
```

```python
folder = '/content/drive/MyDrive/paper/0427/0427-6'
path = '/wireless_.csv'
df = pd.read_csv(folder + path)

wireless_fingerprint = pd.DataFrame(columns =['label', 'Beacon_1', 'Beacon_2', 'Beacon_3', 'Beacon_4', 'Beacon_5', 'Beacon_6', 'Beacon_7'])
min_rssi = [0, 0, 0, 0, 0, 0, 0]
max_rssi = [0, 0, 0, 0, 0, 0, 0]
beacon_rssi_list = {'1':[], '2':[], '3':[], '4':[], '5':[], '6':[], '7':[]}

for i in range(len(df)):
  row = df.loc[i]
  for rssi_idx in range(1, 8):
    beacon_rssi_list[str(rssi_idx)].append(row.iloc[rssi_idx])
```

```python
print(len(beacon_rssi_list['1']))
```

```python
for i in range(1, 8):
  min_rssi[i-1] = min(beacon_rssi_list[str(i)])
  max_rssi[i-1] = max(beacon_rssi_list[str(i)])
  print(min_rssi[i-1],max_rssi[i-1])
```

```python
data = []

for row_idx in range(len(df)):
  rssi = np.zeros(8, dtype=float)
  row = df.loc[row_idx][0:8]
  for rssi_idx in range(1, 8):
    rssi[rssi_idx] = (row.iloc[rssi_idx] - min_rssi[rssi_idx-1]) / (max_rssi[rssi_idx-1] - min_rssi[rssi_idx-1]) * 2 - 1
  rssi[0] = row.iloc[0]
  data.append(rssi)
print(data)
```

```python
columns =['label', 'Beacon_1', 'Beacon_2', 'Beacon_3', 'Beacon_4', 'Beacon_5', 'Beacon_6', 'Beacon_7']
save = pd.DataFrame(data, columns = columns)
save.set_index('label').to_csv(folder + '/normalized.csv')
```
